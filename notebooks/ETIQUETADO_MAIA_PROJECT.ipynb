```python
import pandas as pd

def dividir_articulos(df):
    nuevas_filas = []
    
    for _, row in df.iterrows():
        articulo_id = row['doc_id']
        filename = row['filename']
        texto = row['texto']
        
        # Dividir por párrafos usando doble salto de línea
        parrafos = [p.strip() for p in texto.split("\n\n") if p.strip()]
        
        # Agrupar en bloques de máximo 3 párrafos
        for i in range(0, len(parrafos), 3):
            bloque = "\n\n".join(parrafos[i:i+3])  # conservar doble salto
            nuevas_filas.append({
                'doc_id': articulo_id,
                'filename': filename,
                'sub_id': (i // 3) + 1,
                'texto_bloque': bloque
            })
    
    df_nuevo = pd.DataFrame(nuevas_filas)
    return df_nuevo

# Aplicar a tus DataFrames
Data_I = dividir_articulos(Data_I)
Data_II = dividir_articulos(Data_II)
Data_III = dividir_articulos(Data_III)
Data_IV = dividir_articulos(Data_IV)
Data_V = dividir_articulos(Data_V)
Data_VI = dividir_articulos(Data_VI)
Data_VII = dividir_articulos(Data_VII)
```
